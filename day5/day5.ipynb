```python
def create_maps(src_map):
    maps = src_map.split('\n')
    map_list = []
    for item in maps:
        vals = [int(val) for val in item.split(' ')]
        map_list.append({'source_start':vals[1],'source_end':vals[1]+vals[2],
                            'dest_start':vals[0],'dest_end':vals[0]+vals[2]})
    return pd.DataFrame(map_list)

    
```

```python
def get_dest_value(source_val, df):
    for idx, row in df.iterrows():
        if row.source_start<=source_val<=row.source_end:
            range_diff = source_val -row.source_start
            dest = row.dest_start+range_diff
            return dest

    return source_val
```

```python
from functools import reduce

seeds, *mappings = open('data.txt').read().split('\n\n')
seeds = [int(x) for x in seeds.split()[1:]]

def lookup(inputs, mapping):
    for start, length in inputs:
        while length > 0:
            for m in mapping.split('\n')[1:]:
                dst, src, len = [int(x) for x in m.split()]
                delta = start - src
                if delta in range(len):
                    len = min(len - delta, length)
                    yield (dst + delta, len)
                    start += len
                    length -= len
                    break
            else: yield (start, length); break

print(*[min(reduce(lookup, mappings, s))[0] for s in [
    zip(seeds, [1] * len(seeds)),
    zip(seeds[0::2], seeds[1::2])]])
```
