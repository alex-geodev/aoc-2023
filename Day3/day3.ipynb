```python
import numpy as np
import re
```

```python
rows = input.split('\n')

def get_surrounding_rows(idx):
    current_row = rows[idx]

    if 0<idx<len(rows)-1:
        previous_row = rows[idx-1]
        next_row = rows[idx+1]
        return (previous_row,current_row,next_row)
    elif idx == 0:
        next_row = rows[idx+1]
        return (current_row,next_row)
    else:
        previous_row = rows[idx-1]
        return (previous_row,current_row)
```

```python
def get_number_indexes(row):
    digits = re.finditer(r'\d',row)
    digit_idxs = [digit.start(0) for digit in digits]
    return digit_idxs

def get_symbol_indexes(row):
    symbols = re.finditer(r'[\!\@\#\$\%\^\&\*\-\+\=\/]',row.replace('.','p'))
    symbol_idxs = [symbol.start(0) for symbol in symbols]
    return symbol_idxs
```

```python
def group_numbers(arr):
    groups = []
    sub_group = []
    for idx,num in enumerate(arr):
        sub_group.append(num)
        if idx != len(arr)-1:
            if arr[idx+1]-arr[idx]>1:
                groups.append(sub_group)
                sub_group=[]
        else:
            groups.append(sub_group)
    return groups
```

```python
def get_gear_indexes(row):
    symbols = re.finditer(r'[\*]',row.replace('.','p'))
    symbol_idxs = [symbol.start(0) for symbol in symbols]
    return symbol_idxs
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = ['number','gear_col','gear_row'])
gear_array = []
total=0
row_tracker = 0
for row_number in range(len(rows)):
    subset_rows = get_surrounding_rows(row_number)

    if row_number == 0:
        subset_numbers = get_number_indexes(subset_rows[0])
    else:
        subset_numbers = get_number_indexes(subset_rows[1])
 
    subset_numbers_grouped= group_numbers(subset_numbers)
    
    for num in subset_numbers_grouped:
        if num[0] == 0:
            idx_start = num[0]
        else:
            idx_start = num[0]-1
        idx_end = num[-1]+2
        for idx,subset_row in enumerate(subset_rows):
            #search for symbols in each row based on index range
            symbols = get_gear_indexes(subset_row[idx_start:idx_end])
            if symbols:
                try:
                    part_num = int(subset_rows[0][num[0]:num[-1]+1])
                    df = df._append({'number':part_num, 'gear_col':idx_start+symbols[0], 'gear_row':row_tracker+idx}, ignore_index=True)
                except:
                    part_num = int(subset_rows[1][num[0]:num[-1]+1])                   
                    df = df._append({'number':part_num, 'gear_col':idx_start+symbols[0], 'gear_row':row_tracker-1+idx}, ignore_index=True)

    row_tracker+=1
```
